# Correlations between numeric variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=data.corr(), square=True, annot=True, cmap='crest')
```

- "emp.var.rate", "euribor3m" and "nr.employed" are highly correlated

# Target value bar plot

```python
sns.countplot(data=data, x='y', color='b')
plt.xlabel("Target value")
plt.ylabel("Number of clients")
```

# Age of clients

```python
plt.figure(figsize=(10,7))
sns.histplot(data=data['age'], bins=(data['age'].max() - data['age'].min() + 1))
plt.xlabel('Age')
plt.ylabel('Clients')
plt.title('Histrogram of clients age')
plt.show()
```

# Violin plot for age of clients grouped by target value

```python
sns.violinplot(data=data, x='y', y='age', color='g')
plt.xlabel("Target value")
plt.ylabel("Age")
plt.title("Has client subscribed the deposit?")
plt.show()
```

We can see that in the group that agreed to place the advertised deposit there is a bigger percentage of elderly people than in the other one, so we may assume that they were more likely to use the service.

# Bar plots for categorical variables

```python
plt.figure(figsize=(12,18))

categorical = ['marital','default','housing','loan','contact','poutcome']
titles = ['Marital status of client','Does client have credit in default?','Does client have a housing loan?','Does client have personal loan?','Contact communication type','Previous campaign outcome']

for i, column in enumerate(categorical):
    plt.subplot(3,2,i+1)
    sns.countplot(data=data, x=column, color=(0,0.2,0.6))
    plt.title(titles[i])
    plt.xlabel(' ')
    plt.ylabel("Number of clients")

plt.show()
```

# Education of clients

```python
data['education'].value_counts()
```

```python
import warnings
warnings.filterwarnings("ignore")

education_data = data[['education','y']]
education_data['y'].replace({'yes':1,'no':0}, inplace=True)
education_data.rename({'y':'yes'},axis=1,inplace=True)
education_data['no'] = 1 - education_data['yes']

education_data['education'].replace({'basic.4y' : 'Primary 4 years', 
              'basic.9y' : 'Primary 9 years', 
              'basic.6y' : 'Primary 6 years',
              'high.school' : 'Secondary',
              'professional.course' : 'Professional course',
              'university.degree' : 'Degree', 
              'illiterate' : 'Illiterate', 
              'unknown' : 'Unknown'}, inplace=True)

education_data = education_data[['education','no','yes']].groupby('education').mean()

education_data.plot(kind='bar', figsize=(14,8),stacked=True,
                    xlabel="Education", ylabel="Percentage", rot=0,
                    title="Distribiution of the target value by education")
```

# Quick look into numeric variables

"Duration" column contains information about last contact duration in seconds

```python
data['duration'].describe()
```

```python
data.loc[data['duration'] == 0]
```

```python
sns.boxplot(data=data, x='y', y='duration')
plt.xlabel("Target value")
plt.ylabel("Duration of last contact (in seconds)")
plt.show()
```

We see that duration of last call highly affects the target value. Customers who finally decided to place the advertised deposit spent more time on last contact, as we could have expected. However, if we want our model to be realistic, we shouldn't include this input, beacuse information about how long was the contact is not known before the call is performed and what's more, after the call we already know whether the client agreed to place the deposit.

"Campaign" column contains information about how many times were a client contacted (during this campaign)

```python
data['campaign'].describe()
```

```python
sns.boxplot(data=data, x='y', y='campaign')
plt.xlabel("Target value")
plt.ylabel("Number of contacts")
plt.title("Number of contacts during this campaign")
plt.show()
```

"Previous" column contains information about how many times were a client contacted (before this campaign)

```python
data['previous'].describe()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.countplot(data=data.loc[data['previous'] == 0], x='y', color='b')
plt.xlabel("Target value")
plt.ylabel("Number of clients")
plt.title("Clients not contacted before this campaign")

plt.subplot(1,2,2)
sns.countplot(data=data.loc[data['previous'] > 0], x='y', color='b')
plt.xlabel("Target value")
plt.ylabel("Number of clients")
plt.title("Clients contacted before this campaign")

plt.show()
```

Clients contacted before the current campaign were more likely to place the deposit

"Pdays" column contains information about how many days ago was a client contacted in the previous campaign

```python
data['pdays'].describe()
```

```python
data['pdays'].value_counts().sort_index()
```

999 value in "pdays" column means that the client was not contacted in the previous campaign

```python
data['pdays'].loc[data['pdays'] < 999].describe()
```

```python
data_contacted_before = data.loc[data['pdays'] < 999]
sns.countplot(data=data_contacted_before, x='y', color='g')
plt.xlabel("Target value")
plt.ylabel("Number of clients")
plt.title("Clients contacted in the previous campaign")
```

# Social and economic attributes

```python
plt.figure(figsize=(18,12))

economic_columns = ['emp.var.rate','euribor3m','nr.employed','cons.conf.idx','cons.price.idx']
x_labels = ["Employment vaiation rate","Euribor 3 month rate","Number of employees","Consumer confidence index","Consumer price index"]

for i, column in enumerate(economic_columns):
    plt.subplot(2,3,i+1)
    sns.kdeplot(data=data[column], fill=True)
    plt.xlabel(x_labels[i])
    plt.title(f"Distribution of {x_labels[i]}")

plt.show()
```

# Final observations

- target value distribution is unbalanced (89% 'no' and 11% 'yes')
- some of categorical columns contains missing data ('unknown' values)
- we won't include 'duration' column so that our predictive model is realistic
